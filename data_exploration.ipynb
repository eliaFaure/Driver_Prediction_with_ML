# EXPLORE DATA 

Start by explorig data inside the gps+mpu file. Read both left and right sensors data 

```python
file_path_left = 'archive/PVS 1/dataset_gps_mpu_left.csv'
file_path_right = 'archive/PVS 1/dataset_gps_mpu_right.csv'
```

```python
df_left = pd.read_csv(file_path_left)
df_right = pd.read_csv(file_path_right)
```

```python
#rename the columns 
df_left.columns=['timestamp', 'acc_x_dashboard_l', 'acc_y_dashboard_l', 'acc_z_dashboard_l',
       'acc_x_above_suspension_l', 'acc_y_above_suspension_l',
       'acc_z_above_suspension_l', 'acc_x_below_suspension_l',
       'acc_y_below_suspension_l', 'acc_z_below_suspension_l', 'gyro_x_dashboard_l',
       'gyro_y_dashboard_l', 'gyro_z_dashboard_l', 'gyro_x_above_suspension_l',
       'gyro_y_above_suspension_l', 'gyro_z_above_suspension_l',
       'gyro_x_below_suspension_l', 'gyro_y_below_suspension_l',
       'gyro_z_below_suspension_l', 'mag_x_dashboard_l', 'mag_y_dashboard_l',
       'mag_z_dashboard_l', 'mag_x_above_suspension_l', 'mag_y_above_suspension_l',
       'mag_z_above_suspension_l', 'temp_dashboard_l', 'temp_above_suspension_l',
       'temp_below_suspension_l', 'timestamp_gps', 'latitude', 'longitude',
       'speed']
df_right.columns=[
    'timestamp', 'acc_x_dashboard_r', 'acc_y_dashboard_r', 'acc_z_dashboard_r',
    'acc_x_above_suspension_r', 'acc_y_above_suspension_r', 'acc_z_above_suspension_r',
    'acc_x_below_suspension_r', 'acc_y_below_suspension_r', 'acc_z_below_suspension_r',
    'gyro_x_dashboard_r', 'gyro_y_dashboard_r', 'gyro_z_dashboard_r',
    'gyro_x_above_suspension_r', 'gyro_y_above_suspension_r', 'gyro_z_above_suspension_r',
    'gyro_x_below_suspension_r', 'gyro_y_below_suspension_r', 'gyro_z_below_suspension_r',
    'mag_x_dashboard_r', 'mag_y_dashboard_r', 'mag_z_dashboard_r',
    'mag_x_above_suspension_r', 'mag_y_above_suspension_r', 'mag_z_above_suspension_r',
    'temp_dashboard_r', 'temp_above_suspension_r', 'temp_below_suspension_r',
    'timestamp_gps', 'latitude', 'longitude', 'speed'
]
```

```python
pvs1=pd.concat([df_left,df_right], axis=1)
```

```python
pvs1 = pvs1.loc[:, ~pvs1.columns.duplicated()]
print(f" the new dataframe has {len(pvs1.columns)} columns")
```

Plot the columns 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df_left and df_right have been defined as per your previous code

# List of column pairs for comparison (based on your column naming convention)
columns_to_compare = [
    ('acc_x_dashboard_l', 'acc_x_dashboard_r'),
    ('acc_y_dashboard_l', 'acc_y_dashboard_r'),
    ('acc_z_dashboard_l', 'acc_z_dashboard_r'),
    ('acc_x_above_suspension_l', 'acc_x_above_suspension_r'),
    ('acc_y_above_suspension_l', 'acc_y_above_suspension_r'),
    ('acc_z_above_suspension_l', 'acc_z_above_suspension_r'),
    ('acc_x_below_suspension_l', 'acc_x_below_suspension_r'),
    ('acc_y_below_suspension_l', 'acc_y_below_suspension_r'),
    ('acc_z_below_suspension_l', 'acc_z_below_suspension_r'),
    ('gyro_x_dashboard_l', 'gyro_x_dashboard_r'),
    ('gyro_y_dashboard_l', 'gyro_y_dashboard_r'),
    ('gyro_z_dashboard_l', 'gyro_z_dashboard_r'),
    ('gyro_x_above_suspension_l', 'gyro_x_above_suspension_r'),
    ('gyro_y_above_suspension_l', 'gyro_y_above_suspension_r'),
    ('gyro_z_above_suspension_l', 'gyro_z_above_suspension_r'),
    ('gyro_x_below_suspension_l', 'gyro_x_below_suspension_r'),
    ('gyro_y_below_suspension_l', 'gyro_y_below_suspension_r'),
    ('gyro_z_below_suspension_l', 'gyro_z_below_suspension_r'),
    ('mag_x_dashboard_l', 'mag_x_dashboard_r'),
    ('mag_y_dashboard_l', 'mag_y_dashboard_r'),
    ('mag_z_dashboard_l', 'mag_z_dashboard_r'),
    ('mag_x_above_suspension_l', 'mag_x_above_suspension_r'),
    ('mag_y_above_suspension_l', 'mag_y_above_suspension_r'),
    ('mag_z_above_suspension_l', 'mag_z_above_suspension_r'),
    ('temp_dashboard_l', 'temp_dashboard_r'),
    ('temp_above_suspension_l', 'temp_above_suspension_r'),
    ('temp_below_suspension_l', 'temp_below_suspension_r'),
    ('timestamp_gps', 'timestamp_gps'),  # No need to compare gps coordinates since they're identical
    ('latitude', 'latitude'),            # Same for latitude and longitude
    ('longitude', 'longitude'),
    ('speed', 'speed')
]

# Number of rows and columns for the grid (4x4 grid for 16 pairs)
n_rows = 8
n_cols = 4

# Create the subplots with a larger figure size
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 18))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Loop through the column pairs and assign each plot to a subplot
for i, (col_left, col_right) in enumerate(columns_to_compare):
    if i < len(axes):  # Make sure we don't go out of bounds
        ax = axes[i]  # Get the subplot axis
        ax.plot(df_left[col_left], label=f'{col_left} (Left)', color='blue')
        ax.plot(df_right[col_right], label=f'{col_right} (Right)', color='red')
        
        # Add labels and title for each subplot
        ax.set_xlabel('Index', fontsize=10)
        ax.set_ylabel('Values', fontsize=10)
        ax.set_title(f'{col_left} VS {col_right}', fontsize=10)
        
        # Add legend
        ax.legend(fontsize=8)

# Adjust the layout and spacing to prevent overlap
plt.tight_layout(pad=2.0)  # Increase the padding between subplots
plt.subplots_adjust(hspace=0.5, wspace=0.5)  # Adjust the space between rows and columns

# Show the grid of plots
plt.show()
```
